```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os
```

```python
image_file = 'golden_gate_bridge_sf.jpg'
```

```python
image = Image.open(image_file)
```

If successful, this function returns an [Image](https://pillow.readthedocs.io/en/stable/reference/Image.html#PIL.Image.Image) object. You can now use instance attributes to examine the file contents:

```python
# Display the image
plt.imshow(image)
plt.title("Uploaded Image")
plt.axis('off')
plt.show()
```

```python
# 3. Create NumPy Array from the Image
# Convert the PIL image to a NumPy array
image_array = np.array(image)

print("Image Array Shape:", image_array.shape)
print("Image Array Data Type:", image_array.dtype)
```

```python
# 4. Examine the Contents of the Image Array
# Display some key properties of the array
print("Image Array Shape:", image_array.shape)  # (height, width, channels)
print("Image Array Data Type:", image_array.dtype)  # e.g., uint8
print("Minimum Pixel Value:", np.min(image_array))
print("Maximum Pixel Value:", np.max(image_array))

# Display the first few rows of the array (showing pixel values)
print("\nFirst 5 rows of the image array:")
print(image_array[:5, :5, :])  # Show the first 5x5 pixels for each color channel

# Display the image using matplotlib
plt.imshow(image_array)
plt.title("Image Displayed from NumPy Array")
plt.axis('off')
plt.show()
```
