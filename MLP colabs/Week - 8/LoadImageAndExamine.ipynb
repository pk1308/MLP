# Image Loading and Examination

First of all, download an image of a tourist destination or iconic landmark of your city.

Let's make use of [Pillow](https://pillow.readthedocs.io/en/stable/handbook/tutorial.html) library to read the image.

```python
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
```

```python
# image_path = 'gatewayofindia.jpg'
image_path = 'golden_gate_bridge_sf.jpg'
```

```python
image = Image.open(image_path)
```

```python
plt.imshow(image)
plt.title('Uploaded image')
plt.axis('off')
plt.show()
```

```python
type(image)
```

Convert the image into numpy array

```python
image_array = np.array(image)
```

```python
image_array.shape
```

```python
image_array[:,:,2]
```

```python
image_array.dtype
```

```python
print("Minimum pixel value:", np.min(image_array))
print("Maximum pixel value:", np.max(image_array))
```

```python
# print first 5 rows in each color channel
image_array[:5, :5, :]
```

```python
plt.imshow(image_array)
plt.axis('off')
plt.title('Uploaded image in numpy array format.')
```
