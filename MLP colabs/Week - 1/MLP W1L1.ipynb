No difference - they're identical methods.

isna() and isnull() are aliases of each other in pandas:

df.isna().sum() # Counts missing values per column df.isnull().sum() # Exactly the same result

Both return: Series showing count of missing values for each column.

Convention: isna() is preferred in newer pandas code since it's more explicit about checking for "Not Available" values, but both work identically.

Example: import pandas as pd import numpy as np

df = pd.DataFrame({ 'A': [1, 2, np.nan], 'B': [4, np.nan, 6] })

df.isna().sum() # A: 1, B: 1 df.isnull().sum() # A: 1, B: 1 (same result)

Use whichever you prefer - they're functionally identical!
